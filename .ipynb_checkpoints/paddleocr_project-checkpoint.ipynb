```python
from paddleocr import PaddleOCR,draw_ocr
from matplotlib import pyplot as plt 
import cv2
import os
```

```python
# Import necessary libraries
from paddleocr import PaddleOCR
import cv2
import matplotlib.pyplot as plt

# Initialize the OCR model
ocr = PaddleOCR(use_angle_cls=True, lang="en")

# Path to the image
image_path = os.path.join('.', 'images','10.jpg')  # Change to your image path

# Run OCR on the image
results = ocr.ocr(image_path, cls=True)

# Process and print the results
extracted_text = []
for result in results:
    for line in result:
        text = line[1][0]  # Extract text from the OCR result
        confidence = line[1][1]  # Confidence score
        extracted_text.append(text)
        print(f"Detected: {text} (Confidence: {confidence:.2f})")

# Optionally, display the image with detected words
image = cv2.imread(image_path)
for result in results:
    for line in result:
        points = line[0]  # Coordinates of the bounding box
        x_min, y_min = int(points[0][0]), int(points[0][1])
        x_max, y_max = int(points[2][0]), int(points[2][1])

        # Draw bounding box
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

        # Put text on image
        cv2.putText(image, line[1][0], (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA)

# Show image with detected words
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
